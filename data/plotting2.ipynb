```python
import matplotlib.pyplot as plt

## exp moving average
def EMA(x, y):
    k = 30 / len(x)
    ema_seq = []
    ema_prev = 0
    for t in range(0, len(x)):
        y_t = y[t] * k + ema_prev * (1 - k)
        ema_prev = y_t
        ema_seq.append(y_t)
    return (x, ema_seq)

idx = 20
def new_figure():
    global idx
    idx +=1
    return plt.figure(idx)

line_style=(0, (5, 5))
loc='lower left' 
bbox_to_anchor=(0.9, 0.1)
borderpad=1
```
