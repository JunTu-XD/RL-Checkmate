```python
import matplotlib.pyplot as plt

## exp moving average
def EMA(x, y):
    k = 4 / len(x)
    ema_seq = []
    ema_prev = 0
    for t in range(0, len(x)):
        y_t = y[t] * k + ema_prev * (1 - k)
        ema_prev = y_t
        ema_seq.append(y_t)
    return (x, ema_seq)
```
