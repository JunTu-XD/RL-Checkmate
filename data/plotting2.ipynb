```python
import matplotlib.pyplot as plt

## exp moving average
def EMA(x, y):
    k = 1 / len(x)
    ema_seq = []
    ema_prev = 0
    for t in range(0, len(x)):
        y_t = y[t] * k + ema_prev * (1 - k)
        ema_prev = y_t
        ema_seq.append(y_t)
    return (x, ema_seq)

idx = 20
def new_figure():
    global idx
    idx +=1
    return plt.figure(idx)
```

```python

from data_no_tuning import *

(nt_s_ema_x, nt_s_r_ema) = EMA(no_tuning_SARSA_x_training_seq, no_tuning_SARSA_R_training_seq)
(nt_q_ema_x, nt_q_r_ema) = EMA(no_tuning_Q_Learning_x_training_seq, no_tuning_Q_Learning_R_training_seq)

(_, nt_s_n_ema) = EMA(no_tuning_SARSA_x_training_seq, no_tuning_SARSA_N_training_seq)
(_, nt_q_n_ema) = EMA(no_tuning_Q_Learning_x_training_seq, no_tuning_Q_Learning_N_training_seq)

from data
```

```python
new_figure()

fig, ax  = plt.subplots()

ax.set_xlabel("episodes")
ax2 = ax.twinx()
ax.plot(nt_q_ema_x, nt_q_r_ema, label="R", linestyle="--")
ax.set_ylabel("R")
ax2.plot(nt_q_ema_x, nt_q_n_ema, label="N")
ax2.set_ylabel("N")
fig.legend()
plt.show()
```
